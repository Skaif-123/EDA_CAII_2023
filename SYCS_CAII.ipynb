```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
import numpy as np
df=pd.read_excel("/gdrive/MyDrive/content/PowerLifting.xlsx")
df.head()
```

```python
df['Squat1Kg'].fillna(int(df['Squat1Kg'].mean()),inplace=True)
df['Squat2Kg'].fillna(int(df['Squat2Kg'].mean()),inplace=True)
df['Squat3Kg'].fillna(int(df['Squat3Kg'].mean()),inplace=True)
df['Best3SquatKg'].fillna(int(df['Best3SquatKg'].mean()),inplace=True)

# For BenchPress filling data with means
df['Bench1Kg'].fillna(int(df['Bench1Kg'].mean()),inplace=True)
df['Bench2Kg'].fillna(int(df['Bench2Kg'].mean()),inplace=True)
df['Bench3Kg'].fillna(int(df['Bench3Kg'].mean()),inplace=True)
df['Best3BenchKg'].fillna(int(df['Best3BenchKg'].mean()),inplace=True)

#For DeadLifting filling data with means
df['Deadlift1Kg'].fillna(int(df['Deadlift1Kg'].mean()),inplace=True)
df['Deadlift2Kg'].fillna(int(df['Deadlift2Kg'].mean()),inplace=True)
df['Deadlift3Kg'].fillna(int(df['Deadlift3Kg'].mean()),inplace=True)
df['Best3DeadliftKg'].fillna(int(df['Best3DeadliftKg'].mean()),inplace=True)


#Filling all other details
df=df.fillna(method='bfill')

```

```python
''' 
AS we can see 'Squat4Kg' &'Tested' has no data atall, 
and it will not affect our working with dataset
'''
df=df.drop(['Squat4Kg','Tested','Bench4Kg','Deadlift4Kg'],axis=1)
```
